```python
import numpy as np
```

```python
def gramSchmidt(A):
    n=A.shape[1]
    R = np.zeros((n,n))
    Q = np.zeros(A.shape)
    R[0,0] = np.linalg.norm(A[:,0])
    Q[:,0] = A[:,0]/R[0,0]
    for i in range(1,n):
        Q[:,i] = A[:,i]
        for j in range(0,i):
            R[j,i] = Q[:,j].T @ A[:,i] 
            Q[:,i] = Q[:,i] - R[j,i] * Q[:,j]
        R[i,i] = np.linalg.norm(Q[:,i])
        Q[:,i] = Q[:,i]/R[i,i]
    return Q,R            
```

```python
ep = 1.1e-10
A = np.asarray([[1,1,1],
               [ep,0,0],
               [0,ep,0],
               [0,0,ep]])
```

```python
print(A)
```

```python
Q1,R1 = gramSchmidt(A)
print(Q1)
print('==============================')
print(R1)
```

```python
def modGramSchmidt(A):
    n=A.shape[1]
    R = np.zeros((n,n))
    Q = np.zeros(A.shape)
    R[0,0] = np.linalg.norm(A[:,0])
    Q[:,0] = A[:,0]/R[0,0]
    for i in range(1,n):
        Q[:,i] = A[:,i]
        for j in range(0,i):
            R[j,i] = Q[:,j].T @ Q[:,i] 
            Q[:,i] = Q[:,i] - R[j,i] * Q[:,j]
        R[i,i] = np.linalg.norm(Q[:,i])
        Q[:,i] = Q[:,i]/R[i,i]
    return Q,R    
```

```python
Q2,R2 = modGramSchmidt(A)
print(Q2)
print('==============================')
print(R2)
```
