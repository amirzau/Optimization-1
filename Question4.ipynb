```python
import numpy as np
import scipy.linalg as slin
```

```python
A = np.asarray([[2,1,2],
               [1,-2,1],
               [1,2,3],
               [1,1,1]])

b =np.asarray([[6],
              [1],
              [5],
              [2]])
```

```python
ATA = np.transpose(A) @ A
ATb = np.transpose(A) @ b
```

```python
L = np.linalg.cholesky(ATA)
```

```python
y = slin.solve_triangular(L,ATb,lower=True)
```

```python
x = slin.solve_triangular(np.transpose(L),y,lower=False)
```

```python
U, Sigma ,VT = np.linalg.svd(A)
Sigma_mat = np.zeros(A.shape)
Sigma_mat[:3,:3] = np.diag(Sigma)
```

```python
print(U)
print(Sigma_mat)
print(VT)


```

```python
Sigma_mat_pinv = np.transpose(np.zeros(A.shape))
Sigma_mat_pinv[:3,:3] = np.diag(1/Sigma[:3])
Sigma_mat_pinv.round(3)
```

```python
y = Sigma_mat_pinv @ U.T @ b
```

```python
x = VT.T @ y
```
