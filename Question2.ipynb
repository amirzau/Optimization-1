```python
import numpy as np
```

```python
A = np.asarray([[1,2,3,4],
              [2,4,-4,8],
              [-5,4,1,5],
              [5,0,-3,-7]])
```

```python
A
```

```python
ATA = np.transpose(A) @ A
ATA
```

```python
(value, vector) = np.linalg.eig(ATA)
indexOfMaxValue = np.argmax(value)

print(np.sqrt(value[indexOfMaxValue]))
print(np.linalg.norm(A,2))
```
